```python
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
driver = webdriver.Chrome('/usr/bin/chromedriver', options=options)
```

```python
from tqdm import tqdm
import time
import warnings
warnings.filterwarnings( 'ignore' )
```

```python
brief_url = []

#소아청소년과 https://www.hidoc.co.kr/healthqna/part/list?code=PD000&page=#351
#호흡기내과 https://www.hidoc.co.kr/healthqna/part/list?code=PMP00&page=#624
#감염내과 https://www.hidoc.co.kr/healthqna/part/list?code=PMI00&page=#394
#알레르기내과 https://www.hidoc.co.kr/healthqna/part/list?code=PMA00&page=#39
#이비인후과 https://www.hidoc.co.kr/healthqna/part/list?code=PE000&page=833

for i in tqdm(range(411, 520)):
    url = "https://www.hidoc.co.kr/healthqna/part/list?code=PMI00&page=" + str(i)
    driver.get(url)
    pages_url = driver.find_elements_by_xpath('//*[@id="hidocBody"]/div[1]/div/div[1]/div/a')
    for page in pages_url:
        brief_url.append(page.get_attribute('href'))
```

```python
len(brief_url)
```

```python
urls = []
for url in brief_url:
    urls.append([url])
```

```python
import csv
f = open('hidoc_이비인_url.csv', "w", newline='')
csvfile = csv.writer(f)
for line in urls:
    csvfile.writerow(line)
f.close()
```

```python
def get_text(url):

    driver.get(url)

    title = driver.find_element_by_xpath('//*[@id="hidocBody"]/div[1]/div[1]/div[2]/strong')
    date = driver.find_element_by_xpath('//*[@id="hidocBody"]/div[1]/div[1]/div[2]/div[1]/div[1]/span')
    question = driver.find_element_by_css_selector('#hidocBody > div.health_view > div.box_type1.view_question > div.inner > div.desc > p')
    doctor = driver.find_element_by_xpath('//*[@id="hidocBody"]/div[1]/div[3]/div[2]/div[1]/div[1]/a')
    #clinic = driver.find_element_by_xpath('//*[@id="hidocBody"]/div[1]/div[3]/div[2]/div[1]/div[1]/span')
    answer = driver.find_element_by_css_selector('#hidocBody > div.health_view > div.view_answer > div:nth-child(30) > div.answer_body > div.cont > div:nth-child(1)')

    return [title.text, date.text, question.text, doctor.text, answer.text, url]
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
# url text 가져오기

brief_data = [['title', 'date', 'question', 'doctor', 'answer', 'url']]
for url in tqdm(brief_url):
    if url is not None:
        try:
            data = get_text(url)
        except NoSuchElementException:
            continue
        brief_data.append(data)
        time.sleep(1)
```

```python
import csv

def make_csv(crawl_list):
  file = open("hidoc_감염_1819.csv", 'w', newline='')
  csvfile = csv.writer(file)
  for row in crawl_list :
    csvfile.writerow(row)
  file.close()
```
